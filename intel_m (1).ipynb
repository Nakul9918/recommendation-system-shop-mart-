```python
# Step 5: Predict related products based on purchase history
def predict_related_products(selected_product, selected_category, df_purchases):
    """Predicts top 5 related products based on past purchases in the same category."""

    # Filter past purchases in the selected category
    category_purchases = df_purchases[df_purchases["Category"] == selected_category]

    if category_purchases.empty:
        print(f"\nNo related products found for '{selected_product}' in category '{selected_category}'.")
        return []

    # Count occurrences of other products bought in this category
    product_counts = Counter(category_purchases["Product Name"])  # Now correctly indented
    product_counts.pop(selected_product, None)  # Now correctly indented

    # Get top 5 related products
    sorted_products = [product for product, _ in product_counts.most_common(5)]  # ✅ Fixed indentation

    return sorted_products  # ✅ Fixed indentation

# Assuming df_purchases is your DataFrame of purchases
# Replace this with the actual way you load your purchase data
# For example: df_purchases = pd.read_csv('purchases.csv')
# ... (Code to load df_purchases) ...
import pandas as pd
df_purchases = pd.read_csv('Purchases.csv') # This line was added to load the purchase data into a dataframe called df_purchases
related_products = predict_related_products(selected_product, selected_category, df_purchases)
print(related_products) # or any other operation you need to do with the output
```
